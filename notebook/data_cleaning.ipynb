### dropping irrelevant columns

* Displacement: The engine's displacement, typically measured in cubic centimeters (cc), can affect the amount of fuel burned and, consequently, fuel consumption.

* Engine Type: The type of engine, such as V2, V4, or inline, can influence the combustion efficiency and fuel consumption.

* Fuel System: The type of fuel system, such as injection or carburetion, can impact the fuel-air mixture and, therefore, fuel efficiency.

* Compression Ratio: The compression ratio represents the ratio of the volume of the combustion chamber at the bottom of the piston stroke to the volume at the top. Higher compression ratios can lead to increased efficiency and potentially lower fuel consumption.

* Fuel Control: The method of fuel control, such as desmodromic valve control or electronic fuel injection, can influence the combustion process and fuel efficiency.

* Cooling: The cooling system, such as air or liquid cooling, can affect engine performance and fuel consumption.

* Weight: The weight of the motorcycle can impact its fuel efficiency. Heavier motorcycles may consume more fuel.

* Gearbox and Transmission: The number of gears and the type of transmission (e.g., manual or automatic) can influence engine performance and fuel efficiency.

* Tire Size: The size and type of tires can affect rolling resistance, which influences fuel consumption. 

```python
columns = ['fuel_control','fuel_consumption','fuel_system','torque','power','total_length','total_width','compression','cooling','gearbox','transmission','total_weight','front_tire','rear_tire','engine']

new_df = df[columns]

new_df
```

## DATA CLEANING

```python
new_df.info()
```

```python
new_df.dropna(subset=['fuel_system'],inplace=True,ignore_index=True)
new_df['fuel_system'] = new_df['fuel_system'].apply(lambda x : x.replace('.',' ').split(' ')[0])
```

```python


def clean_fuel_control(df):  # fuction to clean fuel control 
    for i, ele in enumerate(df['fuel_control']):
        if ele == 'Double Overhead Cams/Twin Cam (DOHC)':
            df.at[i, 'fuel_control'] = ele.split('/')[0]
        elif ele in ['Single Overhead Cams (SOHC)', 'Overhead Cams (OHC)', 'Overhead Valves (OHV)']:
            if not pd.isnull(ele):
                df.at[i, 'fuel_control'] = ele.split('(')[0]
    return df

new_df =clean_fuel_control(new_df)
```
